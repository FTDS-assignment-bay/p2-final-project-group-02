```python
import os
import shutil

# Folder tempat file JSON dan folder latihan
source_folder = 'json_file'  # Ganti dengan path folder sumber file JSON Anda
target_folder = 'exercises'  # Folder tempat latihan Anda disimpan

# Mendapatkan semua file di dalam folder sumber
files = os.listdir(source_folder)

# Fungsi untuk memindahkan file JSON
def move_json_files():
    for file_name in files:
        # Cek apakah file adalah file JSON
        if file_name.endswith('.json'):
            # Nama latihan diambil dari nama file JSON (misalnya 'Sit-Up.json' -> 'Sit-Up')
            exercise_name = file_name.replace('.json', '')
            
            # Tentukan target folder latihan berdasarkan nama latihan
            exercise_folder = os.path.join(target_folder, exercise_name)
            
            # Pastikan folder latihan ada, jika tidak buat folder baru
            if not os.path.exists(exercise_folder):
                os.makedirs(exercise_folder)
            
            # Tentukan path sumber dan tujuan
            source_file = os.path.join(source_folder, file_name)
            target_file = os.path.join(exercise_folder, file_name)
            
            # Pindahkan file JSON ke folder yang sesuai
            shutil.move(source_file, target_file)
            print(f"File {file_name} dipindahkan ke {exercise_folder}")

# Jalankan fungsi untuk memindahkan file JSON
move_json_files()
```
